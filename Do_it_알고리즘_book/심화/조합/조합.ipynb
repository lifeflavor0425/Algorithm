# 문제 076 - 이항계수 구하기 1

- n,k 주어졌을 때 이항 계수를 구하는 문제

- D[i][1] = i # i개중 1개를 뽑는 경우의 수는 i개
- D[i][0] = 1 # i개중 1개도 선택하지 않는 경우의 수는 1개
- D[i][i] = 1 # i개중 i개를 선택하는 경우의 수는 1개

```python
n,k = 5, 2
D = [[0 for _ in range(n+1)] for _ in range(n+1)]

for i in range(0,n+1):
    D[i][1] = i
    D[i][0] = 1
    D[i][i] = 1

for i in range(2, n+1):
    for j in range(1,i):
        D[i][j] = D[i-1][j] + D[i-1][j-1] # 조합 점화식

for i in range(0,n+1):
    print(D[i])

print(D[n][k])
```

# 문제 077 - 이항계수 구하기 2

```python
n, k = 5, 2
D = [[0 for _ in range(n+1)] for _ in range(n+1)]
MOD = 10007
for i in range(0, n+1):
    D[i][1] = i 
    D[i][i] = 1 
    D[i][0] = 1 

for i in range(2,n+1):
    for j in range(1,i):
        D[i][j] = (D[i-1][j] + D[i-1][j-1]) % MOD

for i in range(0,n+1):
    print(D[i])

print(D[n][k] % MOD)
```

# 문제 078 - 부녀회장이 될 테야

- a 층의 b호에 사는 조건 : 아래층(a-1)의 1호부터 b호까지 사람들의 수의 합
- k층에 n호에 사는 사람의 수

```python
T = 2
for _ in range(T):
    k = int(input())
    n = int(input())

    D = [[0 for _ in range(n+1)] for _ in range(n+1)]

    for j in range(1,n+1):
        D[j][1] = 1
        D[0][j] = j

    for i in range(1, k+1):
        for j in range(2,n+1):
            D[i][j] = D[i][j-1] + D[i-1][j]
    
    print(D[k][n])

for i in range(0,k+1):
    print(D[i])
```

# 문제 079 - 다리 놓기


- 서쪽 n개, 동쪽 m개 (n<m)
- 서쪽의 사이트 개수 n 만큼 지어려 할 때 다리를 지을 수 있는 경우의 수

```python
T = 3
D = [[0 for _ in range(30+1)] for _ in range(30+1)]

for i in range(0,30+1):
    D[i][1] = i
    D[i][i] = 1
    D[i][0] = 1

for i in range(2,30+1):
    for j in range(1,i+1):
        D[i][j] = D[i-1][j]+D[i-1][j-1]
for _ in range(T):
    n, m = map(int, input().split())
    print(D[m][n])

for i in range(0,30+1):
    print(D[i])
```

# 문제 080 - 조약돌 꺼내기

- 조약돌 n개, 색상 m 종류
- k개 뽑았을 때, 조약돌이 모두 같은 색일 확률

```python
m = int(input())
color_arr = list(map(int, input().split()))
n = sum(color_arr)
k = int(input())

D = [[0 for _ in range(n+1)] for _ in range(n+1)]

for i in range(0,n+1):
    D[i][0] = 1
    D[i][1] = i
    D[i][i] = 1

for i in range(2,n+1):
    for j in range(1,i+1):
        D[i][j] = D[i-1][j] + D[i-1][j-1]

color_sum = 0
for color in color_arr:
    color_sum += D[color][k]

print(color_sum / D[n][k])
```

# 문제 081 - 순열의 순서 구하기

- 1~n 의 경우의 수 n!
- 순서는 영문 사전의 정렬 방식과 유사 -> 오름차순
- 조건 1-> k가 주어지면 k번째 순열을 구한다.
- 조건 2-> 임의의 순열이 주어지면 이 순열이 몇 번째 순열인지 구한다.

```python
n = 4 # 자릿 수
F = [1]*21
S = [0]*21
visited = [False]*21

for i in range(1,n+1): # 팩토리얼 초기화
    F[i] = F[i-1] * i

input_data = list(map(int, input().split()))
if input_data[0] == 1:
    k = input_data[1]
    for i in range(1,n+1):
        cnt = 1
        for j in range(1,n+1):
            if visited[j]:
                continue # 이미 사용한 숫자
            if k <= cnt*F[n-i]:
                k -= (cnt-1)*F[n-i]
                S[i]=j
                visited[j] =True
                break
            cnt += 1
    for i in range(1,n+1):
        print(S[i], end=' ')
else :
    k = 1
    for i in range(1,n+1):
        cnt = 0
        for j in range(1,input_data[i]):
            if not visited[j]:
                cnt += 1
        k += cnt * F[n-i]
        visited[input_data[i]] = True
    print(k)

```

# 문제 082 - 사전 찾기

- 특별한 문자열로 이뤄진 사전 만들기
- 모든 문자열 => n개의 'a'와 m개의 'z'로 이뤄져 잇다 -> 순서 x
- k번째 문자열이 무엇인지 구하는 문제 

```python
n, m, k = 2,2,2
D = [[0 for _ in range(n+m+1)] for _ in range(n+m+1)]

for i in range(0,n+m+1):
    D[i][0]= 1
    D[i][1]= i
    D[i][i]= 1


for i in range(2,n+m+1):
    for j in range(1,i+1):
        D[i][j] = D[i-1][j] + D[i-1][j-1]

for i in range(0,n+m+1):
    print(D[i])

T = n+m
answer = ''
D[n+m][n] # 경우의 수
# a 선택 -> D[3][1] 
while T != 0:
    T -= 1
    # a 선택
    if D[T][m] >= k : # k번째 수 보다 'z'의 총 개수보다 많음
        answer += 'a'
    # z 선택
    else :
        m -= 1
        k -= 1
        answer += 'z'

print(answer)
# aazz
# azaz
# azza

# zaaz
# zaza
# zzaa
```

# 문제 083 - 선물 전달하기

- 선물을 각자 나눠 갖는데, 자신은 본인 선물을 받지 않는다. 경우의 수 구하기

```python
n = 5
D = [0]* 100001
mod = 100000000
D[1]=0
D[2]=1

for i in range(3,n+1): # 한명이 나눠졌을 때 i-1
    D[i] = (i-1)*(D[i-2]+D[i-1]) % mod

print(D[n])
```
